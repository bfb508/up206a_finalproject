## Load Census data

```python
# loading the three data files

gdf_race = gpd.read_file('data/acs2019_5yr_B02001_race_V2.geojson')
```

## Begin to look at the data
Since there are three different datasets, we will need to run commands for each one. 

```python
gdf_race.shape
```

```python
# columns to keep
columns_to_keep = ['geoid',
 'name',
 'B02001001',
 'B02001002',
 'B02001003',
 'B02001004',
 'B02001005',
 'B02001006',
 'B02001007',
 'B02001008',
 'B02001009',
 'B02001010',
 'geometry']
```

```python
# redefine gdf with only columns to keep
gdf_race = gdf_race[columns_to_keep]
```

```python
# check the slimmed down gdf
gdf_race.head()
```

```python
gdf_race.columns = ['geoid',
 'name',
 'Total',
 'White',
 'Black',
 'Native',
 'Asian',
 'Native Hawaiian',
 'Some other race alone',
 'Two or more races',
 'Two races including some other race',
 'Two races excluding some other race, and three or more races',
 'geometry']
```

```python
gdf_race.head()
```

```python
gdf_race['WhitePercentage']=round(((gdf_race['White']/ gdf_race['Total'])*100),2)
gdf_race['BlacPercentage']=round(((gdf_race['Black']/ gdf_race['Total'])*100),2)
gdf_race['NativePercentage']=round(((gdf_race['Native']/ gdf_race['Total'])*100),2)
gdf_race['AsianPercentage']=round(((gdf_race['Asian']/ gdf_race['Total'])*100),2)
gdf_race['HawaiianPercentage']=round(((gdf_race['Native Hawaiian']/ gdf_race['Total'])*100),2)
gdf_race['OtherPercentage']=round(((gdf_race['Some other race alone']/ gdf_race['Total'])*100),2)
gdf_race['TwoPlusPercentage']=round(((gdf_race['Two or more races']/ gdf_race['Total'])*100),2)
gdf_race['TwoInclOtherPercentage']=round(((gdf_race['Two races including some other race']/ gdf_race['Total'])*100),2)
gdf_race['TwoPlusThreePlusPercentage']=round(((gdf_race['Two races excluding some other race, and three or more races']/ gdf_race['Total'])*100),2)

gdf_race.head()
```

```python
gdf_race.head(100).plot(figsize=(10,10),column='White',legend=True)
```

```python
gdf_race.plot(figsize=(12,10),
                 column='Black',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
gdf_race.plot(figsize=(12,10),
                 column='Black',
                 legend=True, 
                 scheme='equal_interval')
```

```python
gdf_race.plot(figsize=(12,10),
                 column='Black',
                 legend=True, 
                 scheme='quantiles')
```

```python
# create the 1x2 subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 12))

# name each subplot
ax1, ax2 = axs

# regular count map on the left
gdf_race.plot(column='Black', 
            cmap='RdYlGn_r', 
            scheme='quantiles',
            k=5, 
            edgecolor='white', 
            linewidth=0., 
            alpha=0.75, 
            ax=ax1, # this assigns the map to the subplot,
            legend=True
           )

ax1.axis("off")
ax1.set_title("Black")

# spatial lag map on the right
gdf_race.plot(column='White', 
            cmap='RdYlGn_r', 
            scheme='quantiles',
            k=5, 
            edgecolor='white', 
            linewidth=0., 
            alpha=0.75, 
            ax=ax2, # this assigns the map to the subplot
            legend=True
           )

ax2.axis("off")
ax2.set_title("White")
```
